# Last challenge
## In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.

```python
data = data[data.NOC == 'USA'][['Edition', 'Discipline', 'Athlete']]
data.head(2)
```

```python
g = data.groupby(['Edition', 'Athlete', 'Discipline'], as_index=False, sort=False).size()
```

```python
g.head()
```

```python
#create list with best medallists
medallists = [value.sort_values('size',ascending=False)[:1] for year, value in g.groupby('Edition')]
```

```python
#list to DataFrame
top = pd.DataFrame()
for i in medallists:
    top = top.append(i)
```
