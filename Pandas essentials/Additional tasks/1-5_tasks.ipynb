## #5. Which of these countries has not hosted the summer Olympics?

```python
hosts = pd.read_csv('list-host-cities-olympic-943j.csv')
hosts.head(1)
```

```python
#dropping unnecessary for analysis columns
hosts = hosts[['City','Country']]
hosts.head(1)
```

```python
#add host country to dataset
olympics = olympics.set_index('City').join(hosts.set_index('City'))
olympics.rename(columns={"Country": "Host"}, inplace = True)
olympics.reset_index(inplace = True)
olympics.head(1)
```

```python
necessary_countries = {'Finland', 'Mexico', 'India', 'Belgium'}
```

```python
necessary_countries.difference(set(olympics.Host))
```
