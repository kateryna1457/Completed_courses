## #8. Which city has held the modern Games three times?

```python
#load data about host countries, cities and years of Olympics
#choose only years which are not present on olympics.csv
#"hosts" dataframe contains error in host city 1948 Olympic game

hosts = pd.read_csv('list-host-cities-olympic-943j.csv')
hosts = hosts[hosts.Year > 2008][['City', 'Year']]
```

```python
#clean data: change dtype of "Year" to be similar with olympics.Edition
#make names of columns equal to olympics DF

hosts = hosts.astype({"City": object , "Year": int})
hosts.rename(columns={'Year': 'Edition'}, inplace = True)
hosts.head(1)
```

```python
#choose necessary columns of olympics DF

olymp = olympics[['City', 'Edition']]
olymp.head(1)
```

```python
#combine two DFs to get total data about years and cities of Olympic games

frames = [hosts, olymp]
all_years = pd.concat(frames)
```

```python
#drop duplicates and data for future Olympic games

all_years = all_years.drop_duplicates()
all_years = all_years[all_years.Edition <= 2021]
```

```python
result = all_years.groupby('City', as_index = False).size().sort_values(by = 'size', ascending = False)
result[result['size'] == 3]
```
