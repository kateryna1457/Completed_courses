```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', skiprows = 4)
```

# Task 1
## Plot the total number of medals awarded at each of the Olympic games throughout history

```python
df.head(3)
```

```python
medals_per_years = df.groupby('Edition').count()['Medal']
medals_per_years
```

```python
result = medals_per_years.plot(kind = 'line', 
                               color = 'green', 
                               title = 'Number of Olympic medals per years')
result.set(xlabel="Year of Olympics", ylabel="Number of medals");
```
