# Task 1
## Plot the total number of medals awarded at each of the Olympic games throughout history.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('olympics.csv', skiprows = 4)
```

```python
df.set_index('Edition', inplace = True)
```

```python
count_medals = df.index.value_counts().sort_index()
count_medals.head()
```

```python
ax = count_medals.plot(kind = 'line', 
                       color = 'navy', 
                       title='Number of medals per years of Olympic games')
ax.set(xlabel="Years of Olympic games", ylabel="Number of medals");
```
