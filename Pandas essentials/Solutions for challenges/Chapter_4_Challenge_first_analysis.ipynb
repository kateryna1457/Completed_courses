```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', skiprows = 4)
```

**Task 1:**
    
    - In which events did Jesse Owens win a medal?

```python
df[df.Athlete.str.contains('OWENS, Jesse')].Event
```

**Task 2:**

    - Which country has won the most men's gold medals in singles badminton over the years?
    
*Sort the results alphabetically by the player's names*

```python
#prepare DataFrame for analysis: choose male athlets with gold medals
#in badminton single competition
gold_male_medalists = df[(df.Medal == 'Gold') 
                         & (df.Gender == 'Men') 
                         & (df.Discipline == 'Badminton') 
                         & (df.Event == 'singles')].sort_values(by = 'Athlete')
```

**Task 3:**
    
    - Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
df_for_analysis = df[(df.Edition >= 1984) & (df.Edition <= 2004)]
```

```python
df_for_analysis[ 'NOC'].value_counts().head(3)
```

**Task 4:**
    
    - Display the male gold medal winners for the 100m track and field sprint event over the years
    
*List the results starting with the most recent*

*Show the Olympic city, edition, athlete, and the county they represent*

```python
#prepare DataFrame for analysis: choose male athlets with gold medals and 100m competition
gold_male_medalists = df[(df.Medal == 'Gold') & (df.Gender == 'Men') & (df.Event == '100m')]
```

```python
gold_male_medalists[['City', 'Edition', 'Athlete', 'NOC']].sort_values(by = 'Edition', ascending = False)
```
