```python
#get sorted set of words from file
words = sorted({word.strip().lower() for word in open('words.txt', 'r')})
```

```python
result = dict()
```

```python
#function to create alphabetically ordered sequence of letters from word

def sample(word):
    return ''.join(sorted(word))
```

```python
#Hypothesis: anagrams consist of only the same letters and their len is equal
#so we need to compare ordered sequence of letters of two words to indentify anagrams

def find_anagrams(myword):
    myword = sample(myword)
    
    for word in words:
        if sample(word) == myword:
            return word
```

```python
result = dict()
```

```python
#create dictionary with sequence of letters as key and set of anagrams from file as value

#we cannot write result[word].add(sample(word)) in loop, 
#as there are many words with the same sequence of letters.
#If we do so, values in dict will be repeated, but not collected by anagrams.
#It will be harder to search for the list of all anagrams by a given word

import collections

result = collections.defaultdict(set)

for word in words:
    result[sample(word)].add(word)
```

```python
#save only those pairs in dict, where there are more than one element in values
#to avoid cases, when a word is the only anagrams of itself

end = {word : sorted(set(wordset)) for word, wordset in result.items() if len(wordset) > 1}
```

```python
#create list with pairs of palindromes
#add to list only those pairs, which were not added yet => word >= word1
#and where word is not a palindrome of itself: no such case as ("aba", "aba") => word != word1

ls = []
for wordset in end.values():
    for word in wordset:
        sample = word[::-1]
        for word1 in wordset:
            if word >= word1 and word1 == sample and word != word1:
                ls.append((word, word1))
```
