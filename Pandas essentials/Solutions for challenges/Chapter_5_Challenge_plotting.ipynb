```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('olympics.csv', skiprows = 4)
```

```python
analysis = df[(df.City == 'Beijing') & (df.Edition == 2008) & (df.NOC == 'CHN')]
```

```python
male = analysis[analysis.Gender == 'Men']
```

```python
female = analysis[analysis.Gender == 'Women']
```

# Task 1:

## Plot the number of medals achieved by the Chinese team (men and woman) in Beijing 2008

```python
analysis.Medal.value_counts().plot(figsize = (5,5), kind = 'pie', colormap = 'viridis')
plt.ylabel('');
```

# Task 2:
## Plot the number of gold, silver and bronze medals for each gender

```python
#number of different types of medals by male
male_bronze = male[male.Medal == 'Bronze'].value_counts()
male_gold = male[male.Medal == 'Gold'].value_counts()
male_silver = male[male.Medal == 'Silver'].value_counts()

male_list = [male_gold.value_counts()[1], 
             male_silver.value_counts()[1], 
             male_bronze.value_counts()[1]]
```

```python
#number of different types of medals by female
female_bronze = female[female.Medal == 'Bronze'].value_counts()
female_gold = female[female.Medal == 'Gold'].value_counts()
female_silver = female[female.Medal == 'Silver'].value_counts()

female_list = [female_gold.value_counts()[1], 
               female_silver.value_counts()[1], 
               female_bronze.value_counts()[1]]
```

```python
w = 0.4
medals = ['Gold', 'Silver', 'Bronze']

bar1 = np.arange(len(analysis.Medal.unique())) #for male bar
bar2 = [i+w for i in bar1] #for female bar

plt.bar(bar1, male_list, w, label = 'Men', color = 'navy')
plt.bar(bar2, female_list, w, label = 'Women', color = 'mediumvioletred')

plt.xlabel('Medals')
plt.ylabel('Gender')
plt.title('Medals per gender')

plt.xticks(bar1 + w/2, medals)

plt.legend(loc = 'upper left')
plt.show()
```
